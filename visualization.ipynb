```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load your dataset
data = pd.read_csv(r"C:\Users\lapi\Documents\LUXDEV\python\practice\orders_trialset.csv")

# Drop unwanted columns
columns_to_exclude = ['order_id', 'postal_code']  # <-- Use exact column names
data_filtered = data.drop(columns=columns_to_exclude, errors='ignore')

# Keep only numeric columns (pairplot works best with numeric data)
numeric_data = data_filtered.select_dtypes(include='number')

# Plot the pairplot
sns.pairplot(numeric_data, corner=True)
plt.suptitle("Pairplot of Selected Variables", y=1.02)
plt.show()
```
